```python

# Import necessary libraries
from manim import *
from IPython.display import Video

# Create a scene
class HelloWorld(Scene):
    def construct(self):

        circle = Circle()
        circle.set_fill(BLUE, opacity=0.5)
        circle.set_stroke(BLUE_E, width=4)
        square = Square()

        self.play(Create(square))
        self.wait()
        
        text = Text("Hello, World-----!")
        self.play(Write(text))
        self.wait(2)

        self.play(ReplacementTransform(square, circle))
        self.wait()
        self.play(circle.animate.stretch(4, 0))
        self.play(Rotate(circle, 90 * DEGREES))
        self.play(circle.animate.shift(2 * RIGHT).scale(0.25))

        text = Text("""
            In general, using the interactive shell
            is very helpful when developing new scenes
        """)
        self.play(Write(text))



        
        control_points = [
            [0, 0, 0], [1, 0, 0], [2, 0, 0],
            [0, 1, 0], [1, 1, 1], [2, 1, 0],
            [0, 2, 0], [1, 2, 0], [2, 2, 0],
        ]

        # Create a mesh for the NURBS surface
        nurbs_surface = Surface(
            lambda u, v: (
                (1-u)*(1-v)*np.array(control_points[0]) +
                u*(1-v)*np.array(control_points[1]) +
                (1-u)*v*np.array(control_points[3]) +
                u*v*np.array(control_points[4])
            ),
            u_range=(0, 2),
            v_range=(0, 2),
            resolution=(10, 10),
        )
        
        # Add the NURBS surface to the scene
        self.add(nurbs_surface)

        self.wait()
        self.wait()


        # Create a surface for the function z = x^2 + y^2
        function_surface = Surface(
            lambda u, v: np.array([u, v, u**2 + v**2]),
            u_range=(-2, 2),
            v_range=(-2, 2),
            resolution=(10, 10),
        )
        
        self.play(ReplacementTransform(nurbs_surface, function_surface))
        function_surface.rotate(PI / 2, axis=RIGHT)
        
        self.wait()
        self.wait()

        




# Render the scene
config.media_width = "75%"
scene = HelloWorld()
scene.render()

# Display the rendered video
Video("media/videos/1080p60/HelloWorld.mp4")
```

```python
# Import necessary libraries from Manim
from manim import *

class HilbertCurve(VMobject):
    CONFIG = {
        "order": 1,
    }

    def __init__(self, **kwargs):
        super().__init__(**kwargs)

        if self.order == 1:
            path_points = [UL, DL, DR, UR]
        else:
            curves = VGroup()
            for i in range(4):
                order = self.order - 1
                curve = HilbertCurve(order=order)

                if i == 0:
                    curve.rotate(PI / 2)
                    curve.flip(axis=RIGHT)
                elif i == 1:
                    curve.rotate(-PI / 2)
                    curve.flip(axis=RIGHT)
                curves.add(curve)
            curves.arrange_in_grid(buff=1)

            # Rearrange curves
            curves = VGroup(curves[0], curves[2], curves[3], curves[1])

            # Taking only corner points and the last point from the curves
            path_points = [
                p for curve in curves for i, p in enumerate(curve.get_points())
                if (i % 3 == 0 or i % (len(curve.get_points()) - 1) == 0)
            ]
        self.set_points_as_corners(path_points)

class HilbertCurveScene(Scene):
    def construct(self):
        n = 8  # Order of the curve
        offset = 1.1  # Offset with respect to height of the camera frame
        colors = VIBGYOR
        frame = self.camera.frame

        old_hilbert = HilbertCurve(order=1)
        old_hilbert.set_height(frame.get_height() / offset)

        final_hilbert = HilbertCurve(order=n)
        colors = color_gradient(colors, len(final_hilbert.get_points()))

        order_is = VGroup(
            Text("Order = ", font_size=48),
            Integer(1, font_size=36)
        )
        order_is.arrange(RIGHT)
        order_is.to_corner(UP + LEFT, buff=0.5)
        self.play(Write(order_is))

        order_tracker = ValueTracker()
        f_always(order_is[1].set_value, order_tracker.get_value)

        for i in range(2, n + 2):
            order_tracker.set_value(i - 1)
            old_hilbert.set_color_by_gradient(colors[:len(old_hilbert.get_points())])

            self.play(
                ShowCreation(old_hilbert),
                run_time=i,
                rate_func=linear
            )
            self.wait()

            if i == n + 1:
                break

            # Removing the old_hilbert
            self.mobjects.pop(1)

            if i == n:
                new_hilbert = final_hilbert
            else:
                new_hilbert = HilbertCurve(order=order_tracker.get_value() + 1)
            new_hilbert.set_height(frame.get_height() / offset)

            self.play(
                old_hilbert.rotate, PI / 2,
                old_hilbert.set_points, new_hilbert.get_points()[:len(old_hilbert.get_points())]
            )

            old_hilbert = new_hilbert

# To render the scene interactively in Jupyter Notebook, run:
# from manim import *
# HilbertCurveScene().render()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def hilbert_curve(order, size=1):
    """Generate the points of a Hilbert curve of a given order."""
    if order == 0:
        return np.array([[0, 0]])
    else:
        points = hilbert_curve(order - 1, size / 2)
        points = np.concatenate([points, points + [0, size / 2], points + [size / 2, size / 2], points + [size / 2, 0]])
        return points

# Generate points for the Hilbert curve of order n
n = 5  # Change this value for different orders
points = hilbert_curve(n)

# Plotting the Hilbert curve
plt.figure(figsize=(8, 8))
plt.plot(points[:, 0], points[:, 1], color='blue')
plt.title(f'Hilbert Curve of Order {n}')
plt.axis('equal')
plt.grid(True)
plt.show()
```
